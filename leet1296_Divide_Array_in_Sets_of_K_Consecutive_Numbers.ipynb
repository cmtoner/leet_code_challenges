# 1296. Divide Array into Sets of K Consecutive Numbers
Medium

Description of the problem is omitted for copyright. It can be found here:

https://leetcode.com/problems/divide-array-in-sets-of-k-consecutive-numbers/


## Explanation
1. Check that $nums.len \mod{k} =0$
  - If no, return False
  - If yes, continue
2. Sort the array
2. While $nums$ is not empty:
    - Get the first number in the array (num)
    - Go from $i=num$ to $num+k$:
        - If $i$ is in the array, remove it
        - If $i$ is not in the array, return False

3. If the loop does not stop early, then the array can be subdivided evenly with each subarray $len=k$.

My solution versus a couple of the top Python solutions posted on the problem page.

```python
import time
import random
import collections

def solution1(nums, k):
  # my solution
    if len(nums)%k==0:
        nums = sorted(nums)
    
        while nums:  
            num = nums[0]
            for i in range(num,num+k,1):
                if i in nums:
                    nums.remove(i)
                else:
                    return(False)
    else:
        return(False)

    return(True)
    
def isPossibleDivide1(arr,k):
    # found here: https://leetcode.com/problems/divide-array-in-sets-of-k-consecutive-numbers/solutions/457625/python-counter-solution/
        count = collections.Counter(arr)
        for n in sorted(count):
            if count[n] > 0:
                need = count[n]
                for i in range(n,n+k):
                    if count[i] < need:
                        return False
                    count[i] -= need
        return True
    
def isPossibleDivide2(A, k):
    #found here: https://leetcode.com/problems/divide-array-in-sets-of-k-consecutive-numbers/solutions/470238/java-c-python-exactly-same-as-846-hand-of-straights/
        c = collections.Counter(A)
        start = collections.deque()
        last_checked, opened = -1, 0
        for i in sorted(c):
            if opened > c[i] or opened > 0 and i > last_checked + 1: return False
            start.append(c[i] - opened)
            last_checked, opened = i, c[i]
            if len(start) == k: opened -= start.popleft()
        return opened == 0

```

## Example 1

Known "True" solution from the problem page.

```python
nums = [1,2,3,3,4,4,5,6]
k = 4

print(f'Soution1: {solution1(nums, k)}')
print(f'isPossibleDivide1: {isPossibleDivide1(nums, k)}')
print(f'isPossibleDivide2: {isPossibleDivide2(nums, k)}')

t0 = time.time()
for _ in range(10**4):
    solution1(nums, k)
t1 = time.time()
print(f'\n\nsolution1 delta_t = {t1-t0}')

t0 = time.time()
for _ in range(10**4):
    isPossibleDivide1(nums,k)
t1 = time.time()
print(f'isPossibleDivide1 delta_t = {t1-t0}')

t0 = time.time()
for _ in range(10**4):
    isPossibleDivide2(nums,k)
t1 = time.time()
print(f'isPossibleDivide2 delta_t = {t1-t0}')
```
