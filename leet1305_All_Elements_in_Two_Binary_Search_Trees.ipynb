```python
import time
import random

def createNode(key):
    # create node in the form of a list: (key/value, left, right, count)
    return([key, [], [], 1])

def insertNode(key, tree=None):
    
    if key==None:
        return(tree)
    
    elif tree==None:
        tree = []

    else:
        if key<tree[0]:
            if tree[1]==[]:
                tree[1] = createNode(key)
            else:
                tree[1] = insertNode(key, tree[1])
        elif key==tree[0]:
            tree[3] += 1
        else:
            if tree[2]==[]:
                tree[2] = createNode(key)
            else:
                tree[2] = insertNode(key, tree[2])

    return(tree)

def createTree(root):
    r = root.copy()
    tree = createNode(r.pop(0))
    
    while r:
        tree = insertNode(r.pop(0),tree)
        
    return(tree)
    

def traverseInorder(tree, path=None):
    
    if path==None:
        path = []
    
    # check that the left subtree is empty: if false, proceed left
    if tree[1]!=[]:
        traverseInorder(tree[1],path)
    
    # append path with current value of leaf 
    for i in range(tree[3]):
        path.append(tree[0])
    
    # check that the right node is empty: if false, proceed right
    if tree[2]!=[]:
        traverseInorder(tree[2],path)
            
    return(path)

def mergeArrays(arr0,arr1):
    # arr0/arr1 are sorted arrays from inorder tree traversal
    
    arr_out = []
    
    # while both arr's not empty
    # compare values and pop minimum
    # add minimum to arr_out
    while arr0 and arr1:
        if arr0[0]<=arr1[0]:
            arr_out.append(arr0.pop(0))
        else:
            arr_out.append(arr1.pop(0))
    
    #catch any leftovers            
    if arr0:
        arr_out = arr_out + arr0
    else:
        arr_out = arr_out + arr1
            
    return(arr_out)

def solution1(r0, r1):
    # create both trees
    # Perform inorder traversal of tree1 and store each node’s value in arr0.
    # Perform inorder traversal of tree2 and store each node’s value in arr1.
    # Combine arr0 and arr1 using merge function of merge sort to create result array.
    # Return result array.

    tree0 = createTree(r0.copy())
    tree1 = createTree(r1.copy())
    
    arr0 = traverseInorder(tree0)
    arr1 = traverseInorder(tree1)
    
    arr_merged = mergeArrays(arr0, arr1)
    
    return(arr_merged)

def solution2(r0,r1):
    """
    Parameters
    ----------
    r0 : list or list like
        linear description of tree 0.
    r1 : list or list like
        linear description of tree 1.

    Returns
    -------
    Merged result.

    """
    tree0 = createTree(r0.copy())
    tree1 = createTree(r1.copy())

    stack0 = []
    stack1 = []

    arr_out = []

    while stack0 or stack1 or tree0 or tree1:
        
        while tree0:
            stack0.insert(0,tree0)
            tree0 = tree0[1]
                    
        while tree1:
            stack1.insert(0,tree1)
            tree1 = tree1[1]
                
        if not(stack1) or (stack0 and stack0[0][0]<=stack1[0][0]):
            tree0 = stack0.pop(0)
            arr_out += [tree0[0] for i in range(tree0[3])] 
            tree0 = tree0[2]
        else:
            tree1 = stack1.pop(0)
            arr_out += [tree1[0] for i in range(tree1[3])] 
            tree1 = tree1[2]
            
    return(arr_out)
```

## Test Case 1


```python
root1 = [2,1,4]
root2 = [1,0,3]
ref_list = [0,1,1,2,3,4]

print(f'solution1==reference: {solution1(root1,root2)==ref_list}')
print(f'solution2==reference: {solution2(root1,root2)==ref_list}')


n_iter = 10**2

t0 = time.time()
for _ in range(n_iter):
    solution1(root1,root2)
t1 = time.time()
print(f'solution1 delta_t= {t1-t0}')


t0 = time.time()
for _ in range(n_iter):
    solution2(root1,root2)
t1 = time.time()
print(f'solution2 delta_t= {t1-t0}')
```
