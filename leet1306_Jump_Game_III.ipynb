```python
import random as r
import time

def solution1(arr, start):
    # breath first search (returns the path)
    if 0 not in arr:
        return(False,[])
    else:
        dests = [n for n in range(len(arr)) if arr[n]==0]
    
    # build path to goal
    stop_cond = False
    dest_reached = False
    paths = [[start]]
    
    while not stop_cond:
        # retrieve first path
        path = paths.pop(0)
        
        # check for duplicate values in path
        # if duplicate found, discard and get next path
        # repeat until non-cyclic path found or paths is empty
        while len(path)!=len(set(path)) and not stop_cond:
            if paths==[]:
                stop_cond = True
            else:
                path = paths.pop(0)
        
        # get valid positions for next path
        # if no valid position found, stop loop
        pos = path[-1]
        next_pos = []
        if pos+arr[pos]<=len(arr)-1 and pos+arr[pos]>=0:
            next_pos.append(pos+arr[pos])
        if pos-arr[pos]<=len(arr)-1 and pos-arr[pos]>=0:
            next_pos.append(pos-arr[pos])
        if next_pos==[]:
            stop_cond = True
    
        # check if path has reached goal
        # else create new path for each position in next_pos    
        if path[-1] in dests:
            dest_reached = True
            stop_cond = True
        else:
            for p in next_pos:
                paths.append(path+[p])
            
    
    return(dest_reached,path)

def solution2(arr,start):
    # BFS (optimized)
    
    if arr[start]==0:
        return(True)
    else:
        visited = set()
        queue = set([start])
        
        while queue-visited:
            # get a node from queue
            v = queue.pop()
            # add node to visited
            visited.add(v)
            # if arr[v]==0, return true
            if arr[v]==0:
                return(True)
            # else get next nodes
            else:
                if v-arr[v]>=0:
                    queue.add(v-arr[v])
                if v+arr[v]<=len(arr)-1:
                    queue.add(v+arr[v])
        
        return(False)
```

## Test Case 1

Result should be TRUE.

```python
# Resulst should be True
nums_arr = [4,2,3,0,3,1,2]
start_pos = 5

solution1(nums_arr, start_pos)
solution2(nums_arr, start_pos)


for s in range(len(nums_arr)):
    print(solution1(nums_arr, s))
    
n_iter = 10**4

t0 = time.time()
for _ in range(n_iter):
    solution1(nums_arr, start_pos)
t1 = time.time()
print(f'solution1 delta_t = {t1-t0}')


t0 = time.time()
for _ in range(n_iter):
    solution2(nums_arr, start_pos)
t1 = time.time()
print(f'solution2 delta_t = {t1-t0}')
```
