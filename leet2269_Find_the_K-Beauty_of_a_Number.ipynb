My solutions (*solution1* and *solution2*) versus a posted solution.

```python
import time

def solution1(num,k):
    # sliding window
    num_str = str(num)
    div = [int(num_str[i:i+k]) for i in range(0,len(num_str)-k+1,1)]
    kb = [1 for d in div if d!=0 and num%d==0]
    return(sum(kb))

def solution2(num,k):
  # sliding window, but faster (in most cases)
    kb_num = 0
    num_str = str(num)

    for i in range(0,len(num_str)-k+1,1):
        test_num = int(num_str[i:i+k])
        if not test_num:
            if num%test_num:
                kb_num += 1
    return(kb_num)

def divisorSubstrings(num,k):
    # found here: https://leetcode.com/problems/find-the-k-beauty-of-a-number/solutions/2100121/easy-python-solution/?orderBy=most_votes&languageTags=python3
    s=str(num)
    c=0
    for i in range(k,len(s)+1):
        x=int(s[i-k:i])
        if x!=0 and num%x==0:
            c+=1
    return c
```

## Test Case 1

```python
num = 240
k = 2
n_iter = 10**6

t0 = time.time()
for _ in range(n_iter):
    solution1(num,k)
t1 = time.time()
print(f'Solution 1: {t1-t0}')

t0 = time.time()
for _ in range(n_iter):
    solution2(num,k)
t1 = time.time()
print(f'Solution 2: {t1-t0}')

t0 = time.time()
for _ in range(n_iter):
    divisorSubstrings(num,k)
t1 = time.time()
print(f'divisorSubstrings: {t1-t0}')
```
