Two solutions: 

1. Sum first $n-1$ elements and add the negative of the sum to the list. This solution is intuitive, but very slow.
2. List of positive and negative integers symmetrical about 0. (Much faster)

```python
import random 
import time

def solution1(n):
    list_out = [i for i in range(n-1)]
    list_out.append(0-sum(list_out))

    return(list_out)

def solution2(n):
        return range(1-n,n,2)
```

## Test Case

```python
# generate array of len n
n = random.randint(1,1000)
solution1(n)
sum(solution1(n))==0

# time each solution for n_iter iterations
n_iter = 10**4
t0 = time.time()
for _ in range(n_iter):
    solution1(10**4)
t1 = time.time()
print(f'solution1 time={t1-t0}')


n_iter = 10**4
t0 = time.time()
for _ in range(n_iter):
    solution2(10**4)
t1 = time.time()
print(f'solution2 time={t1-t0}')
```
