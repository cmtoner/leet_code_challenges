# 1297. Maximum Number of Occurrences of a Substring 

Medium

Description omitted for copyright. Full problem statement here:

https://leetcode.com/problems/maximum-number-of-occurrences-of-a-substring/description/



```python
import random
import collections
import time

def solution1(s, maxLetters, minSize, maxSize):
    """    My solution #1
    Every substring of len minSize is included in every substring of len minSize to len maxSize.
    Therefore we only need to count the substrings of len minSize, with the condition that the number 
    of unique characters is less than or equal to maxLetters.
    """
    res = {}

    for i in range(len(s)-minSize+1):
        sub_str = s[i:i+minSize]
        if len(set(sub_str))<=maxLetters:                
            if sub_str in res.keys():
                res[sub_str] += 1
            else:
                res[sub_str] = 1
    
    if len(res.values())==0:
        return(0)
    else:
        return(max(res.values()))
        
def solution2(s, maxLetters, minSize, maxSize):
    """     My solution #2
    Same idea as solution 1, but using collections (slightly faster in some cases)
    """
    
    sub_str = [s[i:i+minSize] for i in range(len(s)-minSize+1) if len(set(s[i:i+minSize]))<=maxLetters]
    
    if sub_str==[]:
        return(0)
    else:        
        return(collections.Counter(sub_str).most_common(1)[0][1])
    
    
def maxFreq1(s, maxLetters, minSize, maxSize):
    # found here: https://leetcode.com/problems/maximum-number-of-occurrences-of-a-substring/solutions/713509/python-o-n-simple-solution-using-counter/
    # count frequency of characters
    count = collections.Counter()
    
    # Check only for minSize
    for i in range(len(s) - minSize + 1):
        t = s[i : i+minSize]
        if len(set(t)) <= maxLetters:
            count[t] += 1
    
    return max(count.values()) if count else 0
    
def maxFreq2(s, maxLetters, minSize, maxSize):
    # found here: https://leetcode.com/problems/maximum-number-of-occurrences-of-a-substring/solutions/1905801/python-easy-approach/
    s1 = []
    count ={}
    while minSize <= maxSize:
        for i in range(0,len(s)):
            if (i+ minSize) <=len(s) and len(set(s[i: i+ minSize])) <= maxLetters:
                s1.append(s[i: i+ minSize])
        minSize += 1         
    for i in s1:
        count[i] = count[i] + 1 if i in count  else 1      
    return max(count.values()) if count else 0
```

## Example 1.

Input: 
- *s = "aababcaab"*, 
- *maxLetters = 2*, 
- *minSize = 3*, 
- *maxSize = 4*

Output: 2

Explanation: Substring "aab" has 2 ocurrences in the original string.
It satisfies the conditions, 2 unique letters and size 3 (between minSize and maxSize).


```python
s = "aababcaab"
maxLetters = 2
minSize = 3
maxSize = 4

print(f'\n\nSolution1: {solution1(s,maxLetters,minSize,maxSize)}')
print(f'Solution2: {solution2(s,maxLetters,minSize,maxSize)}')
print(f'maxFreq2: {maxFreq1(s,maxLetters,minSize,maxSize)}')
print(f'maxFreq2: {maxFreq2(s,maxLetters,minSize,maxSize)}')

t0 = time.time()
for _ in range(10**4):
    solution1(s, maxLetters, minSize, maxSize)
t1 = time.time()
print(f'\nsolution1 delta_t = {t1-t0}')

t0 = time.time()
for _ in range(10**4):
    solution2(s, maxLetters, minSize, maxSize)
t1 = time.time()
print(f'solution2 delta_t = {t1-t0}')

t0 = time.time()
for _ in range(10**4):
    maxFreq1(s, maxLetters, minSize, maxSize)
t1 = time.time()
print(f'maxFreq1 delta_t = {t1-t0}')

t0 = time.time()
for _ in range(10**4):
    maxFreq2(s, maxLetters, minSize, maxSize)
t1 = time.time()
print(f'maxFreq2 delta_t = {t1-t0}')
```

## Example 2.

Input: 
- *s = "aaaa"*, 
- *maxLetters = 1*, 
- *minSize = 3*, 
- *maxSize = 3*

Output: 2

Explanation: Substring "aaa" occur 2 times in the string. It can overlap.

```python
s = "aaaa" 
maxLetters = 1
minSize = 3
maxSize = 3

print(f'\n\nSolution1: {solution1(s,maxLetters,minSize,maxSize)}')
print(f'Solution2: {solution2(s,maxLetters,minSize,maxSize)}')
print(f'maxFreq2: {maxFreq1(s,maxLetters,minSize,maxSize)}')
print(f'maxFreq2: {maxFreq2(s,maxLetters,minSize,maxSize)}')

t0 = time.time()
for _ in range(10**4):
    solution1(s, maxLetters, minSize, maxSize)
t1 = time.time()
print(f'\nsolution1 delta_t = {t1-t0}')

t0 = time.time()
for _ in range(10**4):
    solution2(s, maxLetters, minSize, maxSize)
t1 = time.time()
print(f'solution2 delta_t = {t1-t0}')

t0 = time.time()
for _ in range(10**4):
    maxFreq1(s, maxLetters, minSize, maxSize)
t1 = time.time()
print(f'maxFreq1 delta_t = {t1-t0}')

t0 = time.time()
for _ in range(10**4):
    maxFreq2(s, maxLetters, minSize, maxSize)
t1 = time.time()
print(f'maxFreq2 delta_t = {t1-t0}')
```
