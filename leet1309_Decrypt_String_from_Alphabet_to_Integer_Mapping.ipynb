# LEET 1309: Decrypt String from Alphabet to Integer Mapping

My solution (*solution1*) versus a few publically posted solutions.

```python
import time 
import random
import re

def solution1(s):
    if 'char_map' not in dir(solution1):
        strings = list('abcdefghijklmnopqrstuvwxyz')
        keys = [str(n) for n in range(1,10,1)] + [str(n)+'#' for n in range(10,27,1)]        
        solution1.char_map = dict(zip(keys,strings))
        solution1.rng = list(solution1.char_map.keys())[-1::-1]
        return(solution1(s))
    else:    
        for k in solution1.rng:
            s = s.replace(k, solution1.char_map[k])
        
    return(s)
        
def freqAlphabets1(s):
    # found here: https://leetcode.com/problems/decrypt-string-from-alphabet-to-integer-mapping/solutions/470770/python-3-two-lines-beats-100-16-ms-with-explanation/
    for i in range(26,0,-1): s = s.replace(str(i)+'#'*(i>9),chr(96+i))
    return s

def freqAlphabets2(s):
    # found here: https://leetcode.com/problems/decrypt-string-from-alphabet-to-integer-mapping/solutions/470867/python-1-line-regex/
        return ''.join(chr(int(i[:2]) + 96) for i in re.findall(r'\d\d#|\d', s))


def freqAlphabets3(s):
    # found here: https://leetcode.com/problems/decrypt-string-from-alphabet-to-integer-mapping/solutions/1217397/python-simple-and-easy-solution-96-faster-help-of-ascii-code/
        ans = ""
        i = len(s)-1
        while i>=0:
            if s[i]=="#":
                ans = ans + chr(int(s[i-2:i])+96)
                i=i-2
            else:
                ans = ans + chr(int(s[i])+96)
            i -= 1
        return ans[::-1]
```

## Test Case 1

```python
s = "10#11#12"
known_solution = "jkab"
print(solution1(s))
print(f'solution1(s) == known_solution\t\t{solution1(s) == known_solution}')

t0 = time.time()
for _ in range(10**4):
    solution1(s)
t1 = time.time()
print(f'\nsolution1 time = {t1-t0}')    

t0 = time.time()
for _ in range(10**4):
    freqAlphabets1(s)
t1 = time.time()
print(f'freqAlphabets1 time = {t1-t0}') 

t0 = time.time()
for _ in range(10**4):
    freqAlphabets2(s)
t1 = time.time()
print(f'freqAlphabets2 time = {t1-t0}') 

t0 = time.time()
for _ in range(10**4):
    freqAlphabets3(s)
t1 = time.time()
print(f'freqAlphabets3 time = {t1-t0}') 
```
